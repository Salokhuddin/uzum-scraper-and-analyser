# Data Cleaning and EDA

1. Data Cleaning

```python
# Importing libraries
import pandas as pd
import numpy as np
import re
```

```python
# Import data
ecom = pd.read_csv('/home/salohuddin/Desktop/pandas/product_details/combined_raw.csv')
```

```python
ecom
```

```python
# Set options
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Check for duplicate values
ecom[ecom.duplicated('link')]
```

```python
# Dropping duplicate values
ecom.drop_duplicates(subset='link', inplace=True)
```

```python
ecom[ecom.duplicated()]
```

```python
# reset index
ecom.reset_index(inplace=True, drop=True)
```

```python
# Remove the domain name from the links to shorten column values and save space
ecom['link'] = ecom['link'].str.replace('https://uzum.uz/uz', '')
```

```python
# Convert 'overall_rating' column into a string (temporarily) to strip whitespace
ecom['overall_rating'] = ecom['overall_rating'].astype(str)
```

```python
# Strip whitespaces in all string columns
for column in ecom.columns:
    ecom[column] = ecom[column].str.strip()
```

```python
# The value 'Yo'q' in 'total_sold' column shows that that item has not sold any, so replace it with zero
ecom.loc[ecom['total_sold'] == "Yoʻq", 'total_sold'] = '0'
```

```python
# Extract the number of 'total_sold' from 'total_sold'
ecom['total_sold'] = ecom['total_sold'].str.replace(r'([^\d])', '', regex=True)
```

```python
# Remove 'so'm' and ' ' in price column
ecom['price'] = ecom['price'].str.replace(" so'm", '')
ecom['price'] = ecom['price'].str.replace(" ", '')
```

```python
# Extract the number of times the item was sold in the last week
ecom['last_week_sale'].str.replace(r'([\d]+)', '', regex=True).unique()
```

```python
def replace_last_week_sale(x):
    if 'haftada' in x:
        return ''.join(re.findall(r'\d', x))
    else:
        return '0'

ecom['last_week_sale'] = ecom['last_week_sale'].apply(replace_last_week_sale)
```

```python
# Extract the number of reviews from 'number_of_reviews'
ecom['number_of_reviews'] = ecom['number_of_reviews'].str.extract(r'([\d]+)')
```

```python
# Replace NaN values with zero since NaN means 'Baholar hali yoʻq'
ecom['number_of_reviews'] = ecom['number_of_reviews'].fillna('0')
```

```python
# Separate 'breadcrumb' into separate columns
ecom[['main_category', 'subcategory', 'subcategory_1', 'subcategory_2', 'subcategory_3']] = ecom['breadcrumb'].str.split('/', expand=True)
```

```python
# Drop 'breadcumb' column, since we do not need it anymore
ecom.drop('breadcrumb', axis=1, inplace=True)
```

```python
ecom['subcategory_2'].unique()
```

```python
ecom['total_sold'].unique()
```

```python
# Convert columns into proper datatypes
ecom[['total_sold', 'last_week_sale', 'number_of_reviews']] = ecom[['total_sold', 'last_week_sale', 'number_of_reviews']].astype('int64')
ecom[['price', 'overall_rating']] = ecom[['price', 'overall_rating']].astype(float)
```

```python
# NaN value in 'subcategory_1', 'subcategory_2','subcategory_3' columns is replaces with '-'
ecom.iloc[:, 10:] = ecom.iloc[:, 10:].fillna('-')
```

```python
ecom.to_csv('/home/salohuddin/Desktop/pandas/product_details/combined_raw_cleaned.csv', index=False)
```

## 2. EDA

```python
# Find the top ten highest-selling products and the number of times each sold
top_ten_products_of_all_time = ecom.sort_values(by='total_sold', ascending=False).head(10)[['product_name', 'total_sold']].set_index('product_name')
```

```python
top_ten_products_of_all_time
```

```python
# Find the top ten products of last week
top_ten_products_of_last_week = ecom.sort_values(by='last_week_sale', 
                                                 ascending=False)[['product_name', 'last_week_sale']].head(10).set_index('product_name')
top_ten_products_of_last_week
```

```python
# Find the worst ten products of last week (among those products sold at least once)
worst_ten_products_of_last_week = ecom[ecom['last_week_sale'] > 0].sort_values(by='last_week_sale', 
                                                 ascending=True)[['product_name', 'last_week_sale']].head(10).set_index('product_name')
worst_ten_products_of_last_week
```

```python
# Categories sorted by the total sales
categories_by_total_sales = ecom.groupby('main_category')['total_sold'].sum().sort_values(ascending=False).reset_index().set_index('main_category')
categories_by_total_sales
```

```python
# Categories sorted by the number of times sold in the last week
categories_by_total_sales = ecom.groupby('main_category')['last_week_sale'].sum().sort_values(ascending=False).reset_index().set_index('main_category')
categories_by_total_sales
```

```python
# top ten merchants by total sales
top_ten_merchants_of_all_time = ecom.groupby('merchant_name')['total_sold'].sum().sort_values(ascending=False).head(10).reset_index().set_index('merchant_name')
top_ten_merchants_of_all_time
```

```python
# top ten merchants by sales in the last week
top_ten_merchants_of_last_week = ecom.groupby('merchant_name')['last_week_sale'].sum().sort_values(ascending=False).head(10).reset_index().set_index('merchant_name')
top_ten_merchants_of_last_week
```

```python
# Number of products across main and sub categories
category_wise_product_count = ecom.groupby(['main_category', 'subcategory'])[['link']].agg('count')
category_wise_product_count.sort_values(by='link', ascending=False)
```

```python
# Number of products sold in each category and subcategory
total_sold_across_categories = ecom.groupby(['main_category', 'subcategory'])[['total_sold']].agg({'total_sold': 'sum'}).sort_values(
    by=['main_category', 'total_sold'], ascending=[True, False])
total_sold_across_categories
```

```python
# most-reviewed products
ecom.sort_values(by='number_of_reviews', ascending=False)[['product_name', 'number_of_reviews', 'overall_rating', 'main_category']].head(10)
```

```python
# Most successful products in each category
ecom.groupby(['main_category'])[['product_name', 'total_sold']].max()
```

```python
# Total sales volume
total_sales = ecom['price'].sum()
print('Total Revenue: {:,.0f} sum'.format(total_sales))
```

```python
# Total number of sold
total_sold = ecom['total_sold'].sum()
print('Total sales: {:,.0f}'.format(total_sold))
```
